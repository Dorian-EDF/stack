```python
import edfg_cube_api as eca
import polars as pl
from great_tables import GT, md, style, loc
from dotenv import load_dotenv

load_dotenv()
```

```python
df=eca.cube_table(["VB"],
                  by=['isin'],
                  date="2026-04-17",
                  niv5="Europe")

print(df)
```

```python
df=eca.cube_table(["Performance VL YTD INS %","VB"],
                  by=['instrument'],
                  date="2026-04-17",
                  niv5="Europe")

df = df.with_columns((pl.col("VB") / pl.col("VB").sum()).alias("Poids"))

df = df.rename({
    "Instrument": "Fonds",
    "VB": "Valorisation",
    "Performance VL YTD INS %": "Perf 2026"
})

df=df.filter(pl.col("Fonds")!="None")

tableau = (
    GT(df).tab_header(
        title =md('Poche Europe'),
        subtitle =md("Valorisation du 17-04-2026")
    )
    .fmt_currency(columns=['Valorisation'],compact=6,currency="EUR",decimals=0)
    .fmt_percent(columns=["Perf 2026","Poids"],decimals=1)
    .tab_source_note(source_note=md("Comgest Compounders sorti totalement du portefeuille depuis le 20/04"))
    # .tab_style(
    #     style=style.fill(color="cyan"),
    #     locations=loc.body(columns="Fonds"))
)

tableau.opt_stylize(style=6,color="blue")
```
