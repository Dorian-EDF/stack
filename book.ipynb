# Attribution de performance relative 2025 — Poche Europe

Méthodologie :

1. Récupère la VB de chaque fonds par ISIN à chaque **date VL** de 2025 (+ dernière VL 2024 pour la borne initiale).
2. Récupère l'historique quotidien de `TOT_RETURN_INDEX_GROSS_DVDS` sur Bloomberg pour chaque fonds (via son `Code Bloomberg`) et pour le benchmark `M7EU Index`.
3. Pour chaque période `[t-1, t]` entre deux dates VL :
   - poids de début de période `w_i,t = VB_i,t-1 / Σ VB_j,t-1`
   - retour du fonds `r_i,t = TRI_t / TRI_{t-1} - 1`
   - retour portefeuille `r_p,t = Σ_i w_i,t · r_i,t`
   - retour benchmark `r_b,t`
4. Chaînage annuel : `R_P = Π(1+r_p,t) - 1`, idem `R_B`.
5. Attribution relative avec **lissage de Carino** :
   - `k_t = (ln(1+r_p,t) − ln(1+r_b,t)) / (r_p,t − r_b,t)`, `K` idem au niveau annuel
   - contribution du fonds `i` : `C_i = Σ_t (k_t/K) · w_i,t · (r_i,t − r_b,t)`
   - garantie : `Σ_i C_i = R_P − R_B`

```python
from edfg_bloomany_api import configure as configure_bloom
from edfg_bloomany_api import get_historical
from edfg_cube_api import configure as configure_cube
import numpy as np

configure_cube(cache_enabled=True)
configure_bloom(cache_enabled=True)

POCHE = "Europe"
BENCHMARK = "M7EU Index"
BLOOM_FIELD = "TOT_RETURN_INDEX_GROSS_DVDS"  # bascule vers "PX_LAST" si besoin

# >>> Parametre principal : changer YEAR pour dupliquer le notebook sur une autre periode
YEAR = 2025
START = f"{YEAR - 1}-12-01"  # inclut decembre de l'annee precedente (borne initiale)
END = f"{YEAR}-12-31"
```

```python
# VB par (date VL, ISIN) pour la poche Europe
# On inclut decembre 2024 pour avoir la derniere VL 2024 = borne initiale du 1er retour 2025

df_vb = (
    eca.cube_table(
        "VB",
        by=["date", "isin"],
        date=(START, END),
        filters={"D_Calendrier[Date VL (Y/N)]": "Y"},
        niv5=POCHE,
    )
    .filter(pl.col("ISIN").is_not_null())
)

print("Dates VL:", sorted(df_vb["Date"].unique().to_list()))
print("Nb ISINs:", df_vb["ISIN"].n_unique())
df_vb

# En cas de filtre vide, decommenter pour verifier les valeurs du flag:
# eca.list_values("D_Calendrier[Date VL (Y/N)]")
```

```python
# Mapping ISIN -> Code Bloomberg + Instrument (pour nommage final)

df_meta = (
    eca.cube_table(
        "VB",
        by=["isin", "instrument", "code_bloomberg"],
        date=(START, END),
        filters={"D_Calendrier[Date VL (Y/N)]": "Y"},
        niv5=POCHE,
    )
    .filter(pl.col("ISIN").is_not_null())
    .unique(subset=["ISIN"])
    .select(["ISIN", "Instrument", "Code Bloomberg"])
)

df_meta
```

```python
# Benchmarks de style disponibles pour les fonds
STYLE_BENCHMARKS = [
    "M7EU Index",       # Broad (default)
    "M7EU000G Index",   # Growth
    "M7EU000V Index",   # Value
    "M7EUSM Index",     # Small / Mid
]

# Mapping manuel : nom d'instrument (colonne Instrument) -> ticker du benchmark de style
# Tout fonds absent du dict tombe sur BENCHMARK (M7EU Index) par defaut
FUND_BENCHMARK: dict[str, str] = {
    # "Nom exact de l'instrument": "M7EU000G Index",
    "ARBEVEL PLUVALCA DISRUPT OPP SI EUR": "M7EUSM Index",
    "AXA AEDIFICANDI": "M7EU000V Index",
    "COMGEST EUROPE COMP EUR EA ACC": "M7EU000G Index",
    "COMGEST GROWTH EUROPE EUR I ACC": "M7EU000G Index",
    "DNCA ARCHER MID CAP EUROPE SI - EUR": "M7EUSM Index",
    "ECHIQ AGENOR IXL EUR NH": "M7EUSM Index",
    "ETF AMUNDI MSCI EUROPE VALUE FACTOR": "M7EU000V Index",
    "ETF ISHARES EUROPE VALUE FACTOR": "M7EU000V Index",
    "ING(L) INVEST EURO HIGH DIV I": "M7EU000V Index",
    "M&G EUROPEAN STGIC VALUE EUR CI ACC": "M7EU000V Index",
    "MANDARINE EUROPE MICROCAP M EUR NH": "M7EUSM Index",
    "ODDO AVENIR EUROPE CI EUR": "M7EUSM Index"
}

# Fonds a ecarter explicitement du calcul d'attribution.
# Leur VB reste dans le cube mais ils ne comptent ni dans r_p ni dans les contributions.
# Les fonds non-matches Bloomberg sont deja ecartes automatiquement ; utiliser cette
# liste pour retirer en plus p.ex. des fonds "techniques" ou hors scope.
EXCLUDED_INSTRUMENTS: set[str] = {
    "EMERGENCE ACTIONS",
    "EMERGENCE ACT 2"
    # "Nom exact de l'instrument a ecarter",
}

# Apercu des fonds a mapper (edite FUND_BENCHMARK / EXCLUDED_INSTRUMENTS puis relance)
print("Fonds -> benchmark de style (defaut = M7EU Index) :")
for inst in sorted(df_meta["Instrument"].unique().to_list()):
    tag = "  [EXCLU]" if inst in EXCLUDED_INSTRUMENTS else ""
    print(f"  {inst:50s} -> {FUND_BENCHMARK.get(inst, BENCHMARK)}{tag}")
```

```python
# Historique Bloomberg quotidien pour tous les fonds + les benchmarks, force en EUR via BQL

from edfg_bloomany_api import get_bql

YELLOW_KEYS = (
    "Equity",
    "Index",
    "Govt",
    "Corp",
    "Curncy",
    "Comdty",
    "Mtge",
    "Muni",
    "Pfd",
)


def normalize_ticker(t: str) -> str:
    if t is None:
        return t
    parts = t.strip().split()
    if parts and parts[-1] in YELLOW_KEYS:
        return " ".join(parts)
    return f"{t.strip()} Equity"


df_meta = df_meta.with_columns(
    pl.col("Code Bloomberg")
    .map_elements(normalize_ticker, return_dtype=pl.Utf8)
    .alias("Code Bloomberg")
)

tickers = df_meta["Code Bloomberg"].drop_nulls().unique().to_list()
all_secs = sorted(set(tickers) | set(STYLE_BENCHMARKS))

bql_field = BLOOM_FIELD.lower()
tickers_str = ", ".join(f"'{t}'" for t in all_secs)

expr = f"""
let(#tri = {bql_field}(currency=EUR);)
get(#tri)
for([{tickers_str}])
with(dates=range({START}, {END}), fill=PREV)
"""

results = get_bql(expr)
df0 = results[0]
tri_col = next(c for c in df0.columns if c not in ("ID", "DATE", "CURRENCY"))

df_bloom = (
    df0.rename({"ID": "Code Bloomberg", "DATE": "Date", tri_col: "TRI"})
    .select(["Code Bloomberg", "Date", "TRI"])
)

missing = set(all_secs) - set(df_bloom["Code Bloomberg"].unique().to_list())
print(
    f"Securities recuperees (BQL, EUR): {df_bloom['Code Bloomberg'].n_unique()} / "
    f"{len(all_secs)}"
)
if missing:
    print(f"Manquantes: {missing}")

df_bloom.head()
```

```python
# Echantillonne les series Bloomberg aux dates VL (ffill si pas de quote le jour exact)
# Le cube renvoie les dates sous forme "2024-12-06T00:00:00" : on parse en Datetime puis on caste en Date
# Idempotent : si Date est deja un pl.Date, on n'y touche pas

if df_vb.schema["Date"] == pl.String:
    df_vb = df_vb.with_columns(
        pl.col("Date").str.to_datetime(strict=False).dt.date().alias("Date")
    ).drop_nulls("Date")

vl_dates = sorted(df_vb["Date"].unique().drop_nulls().to_list())

df_wide = (
    df_bloom.with_columns(pl.col("Date").cast(pl.Date, strict=False))
    .drop_nulls("Date")
    .pivot(values="TRI", index="Date", on="Code Bloomberg")
    .sort("Date")
)

bloom_dates = df_wide["Date"].drop_nulls().to_list()
all_dates = sorted(set(bloom_dates) | set(vl_dates))

df_tri_at_vl = (
    pl.DataFrame({"Date": all_dates}).sort("Date")
    .join(df_wide, on="Date", how="left")
    .fill_null(strategy="forward")
    .filter(pl.col("Date").is_in(vl_dates))
)

print(f"Dates VL retenues: {df_tri_at_vl.height}  (attendu: {len(vl_dates)})")
df_tri_at_vl
```

```python
# Retours entre deux dates VL consecutives : r = TRI_t / TRI_{t-1} - 1
# On ne garde que les periodes dont la *fin* tombe dans l'annee YEAR :
# la 1ere periode retenue est [derniere VL de YEAR-1 -> 1ere VL de YEAR]

import datetime as dt

df_tri_long = (
    df_tri_at_vl.unpivot(index="Date", variable_name="Code Bloomberg", value_name="TRI")
    .sort(["Code Bloomberg", "Date"])
)

df_ret = (
    df_tri_long.with_columns(
        [
            (pl.col("TRI") / pl.col("TRI").shift(1).over("Code Bloomberg") - 1).alias("r"),
            pl.col("Date").shift(1).over("Code Bloomberg").alias("prev_date"),
        ]
    )
    .drop_nulls("r")
    .filter(pl.col("Date") >= dt.date(YEAR, 1, 1))
    .filter(pl.col("Date") <= dt.date(YEAR, 12, 31))
)

print(f"Periodes {YEAR} retenues: de {df_ret['prev_date'].min()} a {df_ret['Date'].max()}")
df_ret
```

```python
# Poids de chaque fonds par date VL : w_i,t = VB_i,t / Σ_j VB_j,t

df_total = df_vb.group_by("Date").agg(pl.col("VB").sum().alias("VB_total"))

df_weights = (
    df_vb.join(df_total, on="Date")
    .join(df_meta.select(["ISIN", "Code Bloomberg"]), on="ISIN")
    .with_columns((pl.col("VB") / pl.col("VB_total")).alias("weight"))
    .select(["Date", "Code Bloomberg", "weight"])
)

df_weights
```

```python
# Assemblage : (prev_date, Date, fonds, r_i, w_i, r_b, bench_ticker, r_bi) par periode
# Les fonds ecartes (non-matches Bloomberg ou listes dans EXCLUDED_INSTRUMENTS) sont retires
# Les poids sont ensuite renormalises pour sommer a 1 sur le perimetre retenu
# -> garantit Σ contrib = R_P − R_B (a la precision numerique)

# Retour du benchmark de poche (r_B)
df_bench = (
    df_ret.filter(pl.col("Code Bloomberg") == BENCHMARK)
    .select(["prev_date", "Date", pl.col("r").alias("r_b")])
)

# Retours de tous les benchmarks de style
df_style_ret = (
    df_ret.filter(pl.col("Code Bloomberg").is_in(STYLE_BENCHMARKS))
    .select(
        [
            "Date",
            pl.col("Code Bloomberg").alias("bench_ticker"),
            pl.col("r").alias("r_bi"),
        ]
    )
)

# Fonds : exclure tous les benchmarks + les instruments ecartes manuellement
df_funds = (
    df_ret.filter(~pl.col("Code Bloomberg").is_in(STYLE_BENCHMARKS))
    .rename({"r": "r_i"})
    .join(df_meta.select(["Code Bloomberg", "Instrument"]), on="Code Bloomberg")
    .filter(~pl.col("Instrument").is_in(list(EXCLUDED_INSTRUMENTS)))
    .with_columns(
        pl.col("Instrument")
        .map_elements(
            lambda n: FUND_BENCHMARK.get(n, BENCHMARK), return_dtype=pl.Utf8
        )
        .alias("bench_ticker")
    )
)

df_attr = (
    df_funds.join(
        df_weights,
        left_on=["prev_date", "Code Bloomberg"],
        right_on=["Date", "Code Bloomberg"],
        how="left",
    )
    .with_columns(pl.col("weight").fill_null(0.0).alias("weight_raw"))
    .drop("weight")
    .join(df_bench.select(["prev_date", "r_b"]), on="prev_date")
    .join(df_style_ret, on=["Date", "bench_ticker"], how="left")
)

# Renormalisation : w_i,t = weight_raw_i,t / Σ_j weight_raw_j,t
df_W_t = df_attr.group_by("Date").agg(pl.col("weight_raw").sum().alias("W_t"))
df_attr = (
    df_attr.join(df_W_t, on="Date")
    .with_columns((pl.col("weight_raw") / pl.col("W_t")).alias("weight"))
    .drop("W_t")
)

# Sanity : Σ poids renormalises == 1, et r_bi present
W_check = df_attr.group_by("Date").agg(pl.col("weight").sum().alias("W")).sort("Date")
print(f"Σ poids renormalises  min={W_check['W'].min():.6f}  max={W_check['W'].max():.6f}")

unmatched = df_attr.filter(pl.col("r_bi").is_null())["Instrument"].unique().to_list()
if unmatched:
    print(f"WARNING : r_bi manquant pour : {unmatched}")

df_attr
```

```python
# Retour portefeuille par periode (r_p = Σ_i w_i * r_i) et chainage annuel
df_port = (
    df_attr.group_by(["prev_date", "Date"])
    .agg((pl.col("weight") * pl.col("r_i")).sum().alias("r_p"))
    .join(df_bench.select(["Date", "r_b"]), on="Date")
    .sort("Date")
)

R_P = float((1 + df_port["r_p"]).product() - 1)
R_B = float((1 + df_port["r_b"]).product() - 1)
print(f"R_P = {R_P:.4%}   R_B = {R_B:.4%}   Excess = {R_P - R_B:.4%}")
df_port
```

```python
# Decomposition Brinson (simplifiee) + lissage Carino
# Par periode :
#   allocation_i,t = (k_t/K) * w_i,t * (r_bi,t - r_B,t)     <- choix de style
#   selection_i,t  = (k_t/K) * w_i,t * (r_i,t  - r_bi,t)    <- choix de fonds
#   contribution_i,t = allocation_i,t + selection_i,t = (k_t/K) * w_i,t * (r_i,t - r_B,t)


def carino_k(rp: np.ndarray, rb: np.ndarray) -> np.ndarray:
    same = np.isclose(rp, rb)
    safe_diff = np.where(same, 1.0, rp - rb)
    return np.where(
        same,
        1.0 / (1.0 + rp),
        (np.log1p(rp) - np.log1p(rb)) / safe_diff,
    )


if np.isclose(R_P, R_B):
    K = 1.0 / (1.0 + R_P)
else:
    K = (np.log1p(R_P) - np.log1p(R_B)) / (R_P - R_B)

df_port = df_port.with_columns(
    pl.Series("k_t", carino_k(df_port["r_p"].to_numpy(), df_port["r_b"].to_numpy()))
)

df_attr2 = df_attr.join(df_port.select(["Date", "k_t"]), on="Date").with_columns(
    [
        (pl.col("k_t") / K * pl.col("weight") * (pl.col("r_bi") - pl.col("r_b")))
        .alias("allocation"),
        (pl.col("k_t") / K * pl.col("weight") * (pl.col("r_i") - pl.col("r_bi")))
        .alias("selection"),
        (pl.col("k_t") / K * pl.col("weight") * (pl.col("r_i") - pl.col("r_b")))
        .alias("contrib_carino"),
    ]
)

df_perf_fund = (
    df_funds.group_by("Code Bloomberg")
    .agg(((1 + pl.col("r_i")).product() - 1).alias("perf_cum"))
)

# Poids moyen affiche = poids brut (part reelle dans la poche, avant renormalisation)
df_poids_moyen = (
    df_attr.group_by("Code Bloomberg")
    .agg(pl.col("weight_raw").mean().alias("weight_avg"))
)

df_bench_map = df_funds.select(["Code Bloomberg", "bench_ticker"]).unique()

df_contrib = (
    df_attr2.group_by("Code Bloomberg")
    .agg(
        [
            pl.col("allocation").sum(),
            pl.col("selection").sum(),
            pl.col("contrib_carino").sum(),
        ]
    )
    .join(df_perf_fund, on="Code Bloomberg")
    .join(df_poids_moyen, on="Code Bloomberg")
    .join(df_bench_map, on="Code Bloomberg")
    .join(df_meta.select(["Code Bloomberg", "Instrument"]), on="Code Bloomberg")
    .with_columns((pl.col("perf_cum") - R_B).alias("perf_relative"))
    .sort("weight_avg", descending=True)
)

print(
    f"Σ allocation = {df_contrib['allocation'].sum():.4%}   "
    f"Σ selection = {df_contrib['selection'].sum():.4%}   "
    f"Σ contribution = {df_contrib['contrib_carino'].sum():.4%}   "
    f"vs Excess = {R_P - R_B:.4%}"
)
df_contrib
```

```python
# --- Diagnostic : d'ou vient l'ecart (Σ contrib − (R_P − R_B)) ?

# Somme des poids matches par periode (Σ_i w_i,t dans df_attr)
df_W = (
    df_attr.group_by(["prev_date", "Date"])
    .agg(pl.col("weight").sum().alias("W_t"))
    .join(df_port.select(["Date", "r_p", "r_b", "k_t"]), on="Date")
    .sort("Date")
)

print("Poids matches par periode (sample) :")
print(df_W.head(10))
print(f"\nW_t moyen : {df_W['W_t'].mean():.4%}")
print(f"W_t min   : {df_W['W_t'].min():.4%}")

# Biais theorique = Σ_t (k_t/K) * r_b,t * (1 - W_t)
biais_theo = float(
    ((df_W["k_t"] / K) * df_W["r_b"] * (1 - df_W["W_t"])).sum()
)
ecart_obs = float(df_contrib["contrib_carino"].sum()) - (R_P - R_B)

print(f"\nEcart observe       Σ contrib − (R_P − R_B) = {ecart_obs:.4%}")
print(f"Biais predit par (1) Σ(k_t/K)·r_b·(1−W_t)    = {biais_theo:.4%}")
print(f"Residuel (source 2 : drop_nulls)             = {ecart_obs - biais_theo:.4%}")


# --- Verification : renormaliser les poids au perimetre matche
# W_t = Σ_matched w_i -> on divise par W_t pour avoir des poids qui somment a 1
df_attr_norm = (
    df_attr.join(df_W.select(["Date", "W_t"]), on="Date")
    .with_columns((pl.col("weight") / pl.col("W_t")).alias("weight"))
)

df_port_norm = (
    df_attr_norm.group_by(["prev_date", "Date"])
    .agg((pl.col("weight") * pl.col("r_i")).sum().alias("r_p"))
    .join(df_bench.select(["Date", "r_b"]), on="Date")
    .sort("Date")
)

R_P_norm = float((1 + df_port_norm["r_p"]).product() - 1)
R_B_norm = float((1 + df_port_norm["r_b"]).product() - 1)

if np.isclose(R_P_norm, R_B_norm):
    K_norm = 1.0 / (1.0 + R_P_norm)
else:
    K_norm = (np.log1p(R_P_norm) - np.log1p(R_B_norm)) / (R_P_norm - R_B_norm)

df_port_norm = df_port_norm.with_columns(
    pl.Series(
        "k_t",
        carino_k(df_port_norm["r_p"].to_numpy(), df_port_norm["r_b"].to_numpy()),
    )
)

contrib_sum_norm = float(
    (
        df_attr_norm.join(df_port_norm.select(["Date", "k_t"]), on="Date")
        .with_columns(
            (
                pl.col("k_t")
                / K_norm
                * pl.col("weight")
                * (pl.col("r_i") - pl.col("r_b"))
            ).alias("c")
        )["c"]
    ).sum()
)

print("\n--- Apres renormalisation des poids ---")
print(f"R_P_norm = {R_P_norm:.4%}   R_B = {R_B_norm:.4%}   Excess = {R_P_norm - R_B_norm:.4%}")
print(f"Σ contrib_norm = {contrib_sum_norm:.4%}")
print(f"Residuel final = {contrib_sum_norm - (R_P_norm - R_B_norm):.6%}")
```
